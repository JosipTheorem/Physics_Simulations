# $\frac{d^2x^i}{d\lambda ^2} + \Gamma^{i}_{jk} \frac{dx^j}{d\lambda} \frac{dx^k}{d\lambda} = 0$
# $\Gamma^{i}_{jk} = 0$
# $\frac{d^2x^i}{d\lambda ^2} = 0$

```python
def func(y,t):
    
    x = y[0]
    x_der = y[1]
    
    return [x_der,0]


lamda = np.linspace(-5,5,100)

x0 = 0 #initial position
x0_der = 0.1 #initial "v"

ini = [x0,x0_der]
#numerical integration
sol=odeint(func,ini,lamda) 
```

```python
plt.plot(lamda,sol[:,1],color='firebrick',ls='--',lw=1.5)
plt.xlabel('t or labmda or affine parameter or whatever')
plt.ylabel('x')
```

```python
plt.plot(lamda,sol[:,0],color='firebrick',ls='--',lw=1.5,label='numericko')
plt.xlabel('lambda')
plt.ylabel('dx/d_lambda')
```
